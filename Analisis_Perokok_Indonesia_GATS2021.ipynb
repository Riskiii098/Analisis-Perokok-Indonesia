# ðŸ“Š Analisis Perokok Indonesia (GATS 2021)

Proyek ini menganalisis data dari GATS 2021 untuk memahami pola merokok dan paparan asap rokok di Indonesia berdasarkan jenis kelamin.  
Visualisasi dibuat menggunakan `pandas`, `matplotlib`, dan data disarikan dari file PDF resmi.

```python
import pdfplumber

with pdfplumber.open("Indonesia National 2021 FactSheet.pdf") as pdf:
    all_text = ""
    for i, page in enumerate(pdf.pages):
        text = page.extract_text()
        if text:
            all_text += f"\n--- Halaman {i+1} ---\n"
            all_text += text

print(all_text)  # tampilkan cuplikan 1000 karakter pertama
```

```python
import pandas as pd

data = {
    "Kategori": [
        "Pengguna tembakau",
        "Perokok",
        "Perokok kretek",
        "Pengguna vape",
        "Usia mulai merokok",
        "Paparan di tempat kerja",
        "Paparan di rumah",
        "Paparan di restoran",
        "Rencana berhenti merokok",
        "Disarankan berhenti oleh nakes"
    ],
    "Pria (%)": [65.5, 64.7, 55.5, 5.8, 15.8, 54.6, 63.7, 82.2, 63.4, 39.4],
    "Wanita (%)": [3.3, 2.3, 1.7, 0.3, 17.4, 29.5, 55.0, 62.1, 64.3, 27.8]
}

df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
x = range(len(df))
bar_width = 0.35

plt.bar(x, df["Pria (%)"], width=bar_width, label="Pria", color='skyblue')
plt.bar([i + bar_width for i in x], df["Wanita (%)"], width=bar_width, label="Wanita", color='salmon')

plt.xticks([i + bar_width/2 for i in x], df["Kategori"], rotation=45, ha="right")
plt.ylabel("Persentase (%)")
plt.title("Perbandingan Perokok dan Paparan Asap Rokok (GATS 2021)")
plt.legend()
plt.tight_layout()
plt.show()
```

## ðŸ“Œ Insight Utama

- **Pria** merupakan perokok dominan: 65.5% pria vs hanya 3.3% wanita.
- Meskipun wanita jarang merokok, **lebih dari 50% wanita terpapar asap rokok di rumah**.
- **Paparan di restoran dan tempat kerja juga tinggi**, terutama pada pria.
- **63% perokok berpikir untuk berhenti**, namun hanya **39% yang disarankan oleh tenaga kesehatan**.
- Edukasi dan layanan berhenti merokok perlu diperkuat untuk menurunkan angka konsumsi tembakau.
